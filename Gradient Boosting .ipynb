# Breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

print("Successful!")
```

```python
breast_cancer_data = load_breast_cancer()

df = pd.DataFrame(data=breast_cancer_data.data,columns= breast_cancer_data.feature_names)
print(df.shape)
```

```python
print(df.head())
```

```python
null_checker = df.isnull().sum()

plt.figure(figsize=(10,9))
sns.heatmap(null_checker,annot=True)
plt.title("Checking for null values")
plt.show()
```

```python
df['target']= breast_cancer_data.target

X = df[breast_cancer_data.feature_names]
y = df['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
# Create Gradient Boosting Classifier
gbc = GradientBoostingClassifier(n_estimators= 100,learning_rate=0.2,random_state=42,max_depth=3)
```

```python
gbc.fit(X_train,y_train)
```

```python
predictions = gbc.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,predictions)*100
print(f"Accuracy score: {accuracy:.2f}%")
```

```python
report = classification_report(y_test,predictions,target_names=breast_cancer_data.target_names)
print("Classification Report:")
print(report)
```
