# Random Forests Classification using Iris Dataset from sklearn

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
%matplotlib inline

print("All Done..")
```

```python
iris = load_iris()

# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column to the DataFrame
df['target'] = iris.target

# Print the feature names
print("Feature names:")
print(iris.feature_names)

# Print the target names
print("\nTarget names:")
print(iris.target_names)

print("\nFirst 5 rows of the dataset:")
print(df.head())

# Total number of rows and columns
print(df.shape)
```

```python
iris_data = load_iris()
X = iris_data.data
y = iris_data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.4)
```

```python
rf= RandomForestClassifier(n_estimators = 20)
rf.fit(X_train,y_train)
```

```python
prediction = rf.predict(X_test)
accuracy = accuracy_score(y_test,prediction)*100
print(f"Accuracy = {accuracy:.2f}%")
```

```python
conf_matrix = confusion_matrix(y_test,prediction)
print(conf_matrix)
```

```python
# Plotting a heatmap to view the confusion matrix
plt.figure(figsize=(10,8))
sns.heatmap(conf_matrix , annot=True,xticklabels=iris_data.target_names, yticklabels=iris_data.target_names)
plt.title("View of the Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Plotting a heatmap for the whole dataset
plt.figure(figsize=(10,18))
sns.heatmap( df, annot=True)
plt.title("View of the Iris Dataset")
plt.xlabel("Features")
plt.ylabel("Target")
plt.show()
```
