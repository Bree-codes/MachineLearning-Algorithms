### As noted the trained data contains all 12 columns including whether the passenger survived or not and the test data is missing 1 column(Survived column).

```python
men = train_data.loc[train_data.Sex =='male']['Survived']
rate_of_men = sum(men)/len(men) * 100

print(f"The Percentage of Men that Survived: {rate_of_men:.3f}%")
```

```python
women = train_data.loc[train_data.Sex =='female']['Survived']
rate_of_women = sum(women)/len(women) * 100

print(f"The Percentage of Women that Survived: {rate_of_women:.3f}%")
```

### Now that we know the survival chances for both men and women
### Lets predict the survival chances for the passengers in the test_data.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score

y = train_data['Survived']

features = ["Pclass", "Sex", "SibSp", "Parch"]

X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = DecisionTreeClassifier(max_depth=3)
model.fit(X,y)

predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('results.csv', index=False)
print("Results were successfully saved!")
```
